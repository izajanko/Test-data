```python
# funkcja wybierająca losową wartość z listy

def random_value_from_list(list):
    return F.array(list)[F.floor(F.rand() * len(list))]
```

```python
# funkcja losująca losowy timestamp 

def random_timestamp(max_diffrence):
    return (F.current_timestamp().cast('long') - F.floor(F.rand() * max_diffrence)).cast('timestamp')
```

```python
# określanie zakresu z jakiego losujemy timestamp

three_hours_in_sec = 3 * 60 * 60
ten_years_in_sec = 10*365 * 24 * 60 * 60
```

```python
# generowanie ciągu znaków

import string
source_characters = string.ascii_letters + string.digits

source_characters
```

```python
# funkcja losująca losowy string z ciągu znaków 

def random_string(max_length):
    return F.concat_ws('', F.slice(F.shuffle(F.split(F.lit(source_characters), '')), 1, max_length))
```

```python
# funkcja generująca randomową liczbę o określonej liczbie cyfr

def random_number(length):
    min_value = 10 ** (length - 1)
    return min_value + F.floor(F.rand() * 9 * min_value)
```

```python
# funkcja generująca randomową liczbę z określonego przedziału

def random_number_min_max(min_value, max_value):
    return min_value + F.floor(F.rand() * (max_value - min_value))
```

```python
partition_number=256

#funkcje do generowania danych

def create_customer(number_of_rows):
    df_customer = (
        spark
        .range(1, number_of_rows, numPartitions=partition_number)
        .select(
            "id",
            random_value_from_list(firstnames).alias("first_name"),
            random_value_from_list(lastnames).alias("last_name"),
            random_number_min_max(18,80).alias("age"),
            random_value_from_list(cities).alias("city"),
            random_timestamp(three_hours_in_sec).alias("last_login_timestamp"),
            random_timestamp(ten_years_in_sec).alias("registration_timestamp"),
            random_string(18).alias("customer_hash"),
            random_string(18).alias("customer_name"),
            F.concat(random_string(12), F.lit("@mail.com")).alias("email"),
            random_number(9).alias("phone_number"),
        )
    )
    return df_customer

def create_store(number_of_rows):
    df_store = (
        spark
        .range(1, number_of_rows, numPartitions=partition_number)
        .select(
            "id",
            F.concat(F.lit("store "), random_string(18)).alias("store_name"),
            random_value_from_list(cities).alias("city"),
            F.concat(F.lit("region "), random_string(5)).alias("region_code"),
            random_value_from_list(countries).alias("country"),
            random_number(12).alias("store_number"),
            random_string(18).alias("store_hash"),
            F.concat(random_string(12), F.lit("@mail.com")).alias("store_email"),
            random_number(9).alias("phone_number"),
            random_number_min_max(1,10).alias("store_ranking")

        )
    )
    return df_store

def create_product(number_of_rows):
    df_product = (
        spark
        .range(1, number_of_rows, numPartitions=partition_number)
        .select(
            "id",
            random_string(30).alias("product_name"),
            random_number(19).alias("product_number"),
            random_value_from_list(colours).alias("product_colour"),
            F.round(random_number_min_max(1,100)+F.rand(),4).alias("product_weight"),
            random_value_from_list(units).alias("product_weight_unit"),
            #F.concat(random_number_min_max(1,500), F.lit("x"), random_number_min_max(1,500), F.lit("x"), random_number_min_max(1,500), F.lit("cm")).alias("product_size"),
            random_number_min_max(1,500).alias("product_height"),
            random_number_min_max(1,500).alias("product_lenght"),
            random_number_min_max(1,500).alias("product_width"),
            random_value_from_list(units_2).alias("product_size_unit"),
            random_timestamp(ten_years_in_sec).alias("sell_start_day"),
            random_timestamp(ten_years_in_sec).alias("sell_end_day"),        
            random_string(8).alias("product_type"),
            random_string(2).alias("model"),
            random_string(10).alias("product_category_name"),
            random_number(10).alias("product_category_number"),
        )
    )
    return df_product

def create_employee(number_of_rows):
    df_employee = (
        spark
        .range(1, number_of_rows, numPartitions=partition_number)
        .select(
            "id",
            random_value_from_list(firstnames).alias("first_name"),
            random_value_from_list(lastnames).alias("last_name"),
            random_number_min_max(18,80).alias("age"),
            random_timestamp(ten_years_in_sec).alias("hire_date"),
            random_string(18).alias("employee_identification"),
            F.concat(random_string(12), F.lit("@mail.com")).alias("email"),
            random_number(9).alias("phone_number"),
            random_number(19).alias("account_number"),
            F.round(random_number_min_max(1000,100000),-3).alias("salary"),
        )
    )
    return df_employee

def create_sale(number_of_rows):
    df_sale = (
        spark
        .range(1, number_of_rows, numPartitions=partition_number)
        .select(
            "id",
            random_number_min_max(1,1_000_000_000).alias("product_id"),
            random_number_min_max(1,1_000_000_000).alias("customer_id"),
            random_number_min_max(1,1_000_000_000).alias("employee_id"),
            random_number_min_max(1,1_000_000_000).alias("store_id"),
            random_timestamp(ten_years_in_sec).alias("order_date"),
            random_timestamp(ten_years_in_sec).alias("shipping_date"),
            random_number(10).alias("purchase_order_number"),
            random_value_from_list(payment_methods).alias("payment_method"),
            F.round(random_number_min_max(10,1_000_000_000),-3).alias("unit_price"),
            random_value_from_list(currency).alias("currency"),
            random_number_min_max(1,1000).alias("quantity"),
        )
    )
    return df_sale
```

```python
# funcje do określenia ile GB zajmują dane datasety

#display the files of the current directory & subdirectory recursively in Databricks file system
def get_dir_content(ls_path):
    dir_paths = dbutils.fs.ls(ls_path)
    subdir_paths = [get_dir_content(p.path) for p in dir_paths if p.isDir() and p.path != ls_path]
    flat_subdir_paths = [p for subdir in subdir_paths for p in subdir]
    return dir_paths + flat_subdir_paths


def get_file_sizes(container_name, catlog_name):
    return(
        spark
            .createDataFrame(get_dir_content(f"abfss://{container_name}@storageaccount01ij.dfs.core.windows.net/{catlog_name}/"))
            .filter(F.col("path").endswith(".parquet")) #wybieramy tylko pliki parquet i te, które nie są puste
            .filter(F.col("size") > 0)
            .select(
                "path",
                "name",
                (F.col("size") / 1024.0 / 1024.0 / 1024.0).alias('size_in_gb') #zamiana na GB
            )
    )
    
# get_dir_content(f"abfss://random-data@storageaccount01ij.dfs.core.windows.net/dataset-1gb/")
```

```python
# rozmiar całego katalogu
def sum_size(df):
    return df.groupBy().sum()
# rozmiar całego katalogu z zaokrągleniem do 2 miejsc po przecinku
def sum_with_round(df):
    return df.groupBy().sum().select(F.round(F.col("sum(size_in_gb)"),2))
```
